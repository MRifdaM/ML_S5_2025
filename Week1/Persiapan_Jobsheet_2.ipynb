```python


import os, re, math
import pandas as pd
import numpy as np

# --- Load dari upload ---
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Helper cari kolom by alias (tahan variasi nama)
def find_col(df, aliases):
    cols = {c.strip().lower(): c for c in df.columns}
    for a in aliases:
        key = a.strip().lower()
        if key in cols:
            return cols[key]
    # fallback: buang spasi & tanda baca utk pencocokan longgar
    norm = {re.sub(r"[\s\W]+", "", c.lower()): c for c in df.columns}
    for a in aliases:
        a_norm = re.sub(r"[\s\W]+", "", a.lower())
        if a_norm in norm:
            return norm[a_norm]
    return None

# --- Fungsi pembersihan ---
def normalize_gender(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().lower()
    s_norm = re.sub(r"[\s\.\-_]+", "", s)
    laki = {"l","lk","laki","lakilaki","pria","cowok","male","m"}
    perempuan = {"p","pr","perempuan","wanita","cewek","female","f"}
    if s_norm in laki or "laki" in s_norm: return "Laki - Laki"
    if s_norm in perempuan or "perempuan" in s_norm or "wanita" in s_norm: return "Perempuan"
    if "laki" in s: return "Laki - Laki"
    if "perempuan" in s or "wanita" in s: return "Perempuan"
    return None

def clean_kota(val):
    if pd.isna(val): return np.nan
    s = str(val).strip()
    s = re.sub(r"^\s*(kota\s+administrasi\s+|kota\s+adm\.\s+|kota\s+|kab(?:\.|upaten)\s+)", "", s, flags=re.IGNORECASE)
    s = s.strip(" .,/\\-")
    out = s.title()
    out = re.sub(r"\bDki\b", "DKI", out)
    out = re.sub(r"\bDiy\b", "DIY", out)
    return out

def to_float(val):
    if pd.isna(val): return np.nan
    s = str(val)
    m = re.search(r"[-+]?\d+(?:[.,]\d+)?", s)
    if not m: return np.nan
    num = m.group(0).replace(",", ".")
    try: return float(num)
    except: return np.nan

def clean_kelas(val):
    if pd.isna(val): return np.nan
    s = str(val).strip()
    s = re.sub(r"^(TI|MI|TRPL|SIB|IF|TK|TE|RPL|Teknik\s*Informatika|Manajemen\s*Informatika)\s*[-:]\s*", "", s, flags=re.IGNORECASE)
    if "-" in s: s = s.split("-")[-1]
    s = re.sub(r"^(TI|MI|TRPL|SIB|IF|TK|TE|RPL)\s*", "", s, flags=re.IGNORECASE)
    return s.strip()

# --- Temukan kolom target (pakai alias biar fleksibel) ---
col_gender     = find_col(df, ["Kelamin","Jenis Kelamin","Gender"])
col_kota       = find_col(df, ["Asal Kota","Kota Asal","Kota"])
col_berat      = find_col(df, ["Berat Badan","Berat (kg)","Berat"])
col_tinggi     = find_col(df, ["Tinggi Badan","Tinggi (cm)","Tinggi"])
col_kelas      = find_col(df, ["Asal Kelas (Kelas semester sebelumnya)","Asal Kelas","Kelas semester sebelumnya"])
col_target_ips = find_col(df, ["Target IPS (target realistis IPS semester ini)","Target IPS","IPS Target"])

# --- Terapkan pembersihan sesuai permintaan ---
if col_gender is not None:
    df[col_gender] = df[col_gender].map(normalize_gender)

if col_kota is not None:
    df[col_kota] = df[col_kota].map(clean_kota)

if col_berat is not None:
    df[col_berat] = df[col_berat].map(to_float)

if col_tinggi is not None:
    df[col_tinggi] = df[col_tinggi].map(to_float)

if col_kelas is not None:
    df[col_kelas] = df[col_kelas].map(clean_kelas)

if col_target_ips is not None:
    df[col_target_ips] = df[col_target_ips].map(to_float)

# --- Simpan hasil ---
clean_name = filename.rsplit(".", 1)[0] + "_cleaned.csv"
df.to_csv(clean_name, index=False)
print("Saved:", clean_name)

display(df.head(30))
```

```python
# === NORMALISASI (Min-Max) — versi robust ===
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import re

def find_col(df, aliases):
    cols = {c.strip().lower(): c for c in df.columns}
    for a in aliases:
        key = a.strip().lower()
        if key in cols:
            return cols[key]
    norm = {re.sub(r"[\s\W]+", "", c.lower()): c for c in df.columns}
    for a in aliases:
        a_norm = re.sub(r"[\s\W]+", "", a.lower())
        if a_norm in norm:
            return norm[a_norm]
    return None

# Deteksi ulang kolom penting
col_berat = find_col(df, ["Berat Badan", "Berat (kg)", "Berat"])
col_tinggi = find_col(df, ["Tinggi Badan", "Tinggi (cm)", "Tinggi"])
col_target_ips = find_col(df, ["Target IPS (target realistis IPS semester ini)", "Target IPS", "IPS Target"])

# Kumpulkan kolom numerik yang akan di-scale
num_cols = [c for c in [col_berat, col_tinggi, col_target_ips] if c is not None and c in df.columns]

# Paksa numeric (kalau masih object) agar bisa di-scale
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Opsi: exclude target jika itu label
EXCLUDE_TARGET_AS_LABEL = False  # ubah ke True jika Target IPS adalah label
if EXCLUDE_TARGET_AS_LABEL and col_target_ips in num_cols:
    num_cols = [c for c in num_cols if c != col_target_ips]

# Fallback tambahan: gabungkan dengan semua kolom bertipe numerik (jika perlu)
all_numeric = df.select_dtypes(include=np.number).columns.tolist()
for c in all_numeric:
    if c not in num_cols:
        num_cols.append(c)

# Hindari kolom ID/indeks jika ada (opsional)
# num_cols = [c for c in num_cols if c.lower() not in {"id","index"}]

print("Kolom numerik yang akan di-Normalisasi:", num_cols)

scaler = MinMaxScaler()
df_norm = df.copy()
df_norm[num_cols] = scaler.fit_transform(df_norm[num_cols])

df_norm.to_csv("normalized_output.csv", index=False)
print("Saved: normalized_output.csv")

display(df_norm.head(30))
```

```python
# === STANDARISASI (Z-score) — versi robust ===
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import re

def find_col(df, aliases):
    cols = {c.strip().lower(): c for c in df.columns}
    for a in aliases:
        key = a.strip().lower()
        if key in cols:
            return cols[key]
    norm = {re.sub(r"[\s\W]+", "", c.lower()): c for c in df.columns}
    for a in aliases:
        a_norm = re.sub(r"[\s\W]+", "", a.lower())
        if a_norm in norm:
            return norm[a_norm]
    return None

col_berat = find_col(df, ["Berat Badan", "Berat (kg)", "Berat"])
col_tinggi = find_col(df, ["Tinggi Badan", "Tinggi (cm)", "Tinggi"])
col_target_ips = find_col(df, ["Target IPS (target realistis IPS semester ini)", "Target IPS", "IPS Target"])

num_cols = [c for c in [col_berat, col_tinggi, col_target_ips] if c is not None and c in df.columns]

for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

EXCLUDE_TARGET_AS_LABEL = False  # ubah ke True jika Target IPS adalah label
if EXCLUDE_TARGET_AS_LABEL and col_target_ips in num_cols:
    num_cols = [c for c in num_cols if c != col_target_ips]

all_numeric = df.select_dtypes(include=np.number).columns.tolist()
for c in all_numeric:
    if c not in num_cols:
        num_cols.append(c)

print("Kolom numerik yang akan di-Standarisasi:", num_cols)

scaler = StandardScaler()
df_std = df.copy()
df_std[num_cols] = scaler.fit_transform(df_std[num_cols])

df_std.to_csv("standardized_output.csv", index=False)
print("Saved: standardized_output.csv")


display(df_std.head(30))
```

```python
# === ENCODING (One-Hot) ===
import pandas as pd

# Tentukan kolom kategorikal yang perlu di-encode
cat_cols = []
for c in [col_gender, col_kota, col_kelas]:
    if c is not None and c in df.columns:
        cat_cols.append(c)

# One-Hot Encoding (menjaga semua kategori, handle_unknown implicit via get_dummies)
df_encoded = pd.get_dummies(
    df,
    columns=cat_cols,
    drop_first=False,   # set True jika ingin menghindari multikolinearitas
    dtype=int
)

print("One-Hot Encoding pada kolom:", cat_cols)
df_encoded.to_csv("encoded_output.csv", index=False)
print("Saved: encoded_output.csv")

display(df_encoded.head(30))
```

```python
# --- Import Library ---
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from google.colab import files

# --- Load Dataset ---
filename = list(uploaded.keys())[0]   # Ambil nama file yang diupload
df = pd.read_csv(filename)

# --- Lihat 5 data teratas ---
print("Preview Data:")
display(df.head())

# 1. Pisahkan variabel yang digunakan & tidak digunakan
# Variabel yang tidak digunakan
not_used = []
if 'id' in df.columns:
    not_used.append('id')
if 'Unnamed: 32' in df.columns:
    not_used.append('Unnamed: 32')

# Variabel yang digunakan (selain yang tidak digunakan)
used = [col for col in df.columns if col not in not_used]

print("\nVariabel yang TIDAK digunakan:", not_used)
print("\nVariabel yang DIGUNAKAN:", used)

# Buat dataframe baru hanya dengan variabel yang digunakan
df = df[used]

# 2. Encoding kolom diagnosis (M=1, B=0)

le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

print("\nUnique value diagnosis setelah encoding:", df['diagnosis'].unique())
display(df.head(100))
# 3. Standardisasi semua kolom numerik
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\nData setelah standardisasi:")
display(df.head())
```
